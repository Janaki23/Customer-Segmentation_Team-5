```python
from sklearn.cluster import AgglomerativeClustering

# Specifying the number of clusters
n_clusters = 5

# Performing hierarchical clustering with the optimal number of clusters
model = AgglomerativeClustering(n_clusters=n_clusters, metric='euclidean', linkage='ward')

dropped_Columns= ['CustomerID', 'Location']
# Drop the specified columns
data_for_clustering = mall_Data.drop(dropped_Columns, axis=1)
clusters = model.fit_predict(mall_Data.drop(dropped_Columns, axis=1))

# Adding the cluster labels to the original dataset
mall_Data['Cluster'] = clusters

# Visualizing the clustering results
plt.figure(figsize=(10, 6))
for cluster in range(n_clusters):
    cluster_data = mall_Data[mall_Data['Cluster'] == cluster]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'],
                label=f'Cluster {cluster}', alpha=0.7)
plt.title('Hierarchical Clustering Results')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Loading the dataset
data = pd.read_csv('Mall_Customers_Updated.csv')

# Selecting numeric columns for clustering
numeric_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Items_Purchased_Monthly']
X = data[numeric_cols]

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Performing the hierarchical clustering
linkage_matrix = linkage(X_scaled, method='ward', metric='euclidean')

# Ploting the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, labels=data['CustomerID'].values, leaf_rotation=90, leaf_font_size=8)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Customer ID')
plt.ylabel('Distance')
plt.show()
```

```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Initializing the Dash app
app = dash.Dash(__name__)

# Defining the layout of the dashboard
app.layout = html.Div([
    html.H1("Customer Segmentation Dashboard"),
    
    # Dropdown for selecting clustering variable
    html.Label("Select clustering variable:"),
    dcc.Dropdown(
        id='clustering-variable',
        options=[
            {'label': 'Age', 'value': 'Age'},
            {'label': 'Annual Income (k$)', 'value': 'Annual Income (k$)'},
            {'label': 'Spending Score (1-100)', 'value': 'Spending Score (1-100)'},
            {'label': 'Location', 'value': 'Location'},
            {'label': 'Items Purchased Monthly', 'value': 'Items_Purchased_Monthly'}
        ],
        value='Age'
    ),
    
    # Scattering plot for visualizing clusters
    dcc.Graph(id='cluster-scatterplot'),
])

# Defining callback to update scatter plot based on clustering variable selection
@app.callback(
    dash.dependencies.Output('cluster-scatterplot', 'figure'),
    [dash.dependencies.Input('clustering-variable', 'value')]
)
def update_scatterplot(selected_variable):
    # Creating scatter plot with clusters based on selected variable
    fig = px.scatter(mall_Data, x='CustomerID', y=selected_variable, color='Cluster',
                     title=f'Clustered by {selected_variable}',
                     labels={'CustomerID': 'Customer ID', selected_variable: selected_variable})
    return fig

# Running the app
if __name__ == '__main__':
    app.run_server(debug=True, port=8088)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder

# Loading the dataset
data = pd.read_csv("Mall_Customers_Updated.csv")  # Replace "your_dataset.csv" with the actual filename

# Separating features (X) and target variable (y)
X = data.drop(columns=['CustomerID'])  # Assuming CustomerID is not a predictive feature
y = data['Items_Purchased_Monthly']

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X)

# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initializing regression models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regression': RandomForestRegressor(n_estimators=100, random_state=42),
    'Decision Tree Regression': DecisionTreeRegressor(random_state=42),
    'Gradient Boosting Regression': GradientBoostingRegressor(random_state=42)
}

# Training and evaluating each model
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MSE': mse, 'R^2 Score': r2}

# Printing the results
print("Regression Model Performance:")
for name, result in results.items():
    print(f"{name}:")
    print(f"  Mean Squared Error: {result['MSE']}")
    print(f"  R^2 Score: {result['R^2 Score']}")
    print()
```
